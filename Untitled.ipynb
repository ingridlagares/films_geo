```python
import re
import requests
```

```python
directores = []
with open('film.dbpedia.graph') as dbpedia:
  for linha in dbpedia:
    tokens = linha.split()
    if tokens[1] == '<director>':
      director = tokens[2]
      if director not in directores:
        directores.append(director.replace('<','').replace('>',''))
```

```python
def get_location(name):
    API = 'https://en.wikipedia.org/w/api.php?action=query&prop=revisions&rvprop=content&format=json&titles=%s'

    try:
        req = requests.get(API % name)

        req_json = req.json()
        content = req_json["query"]["pages"].values()[0]['revisions'][0]['*']
        regex_result = re.search(r'birth_place\s*=\s*([^|]+)', content)
        location = regex_result.groups(1)[0].replace('[', '').replace(']', '')
        return location
    except:
        return None
```

```python
def get_gps(location):
    API = 'https://nominatim.openstreetmap.org/search?q=%s&format=json&addressdetails=1'
    try:
        requests.get(API % location)
    
        req_json = req.json()
        lat = req_json[0]['lat']
        lon = req_json[0]['lon']
        
        return (lat, lon)
    except:
        return None
```
